```python
import shutil, os

import numpy as np
import cv2
import skimage
```

```python
from skimage.transform import hough_line, hough_line_peaks
from skimage.feature import canny
from skimage import data
from skimage.transform import probabilistic_hough_line
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def func1(image_path):
    image = cv2.imread(image_path)
    # smoothen the image
    enhanced_image = cv2.medianBlur(image, 5)
    
    #denoise the image
    dst = cv2.fastNlMeansDenoisingColored(enhanced_image, None, 5, 0,0,0)
    
    gray1 = cv2.cvtColor(dst, cv2.COLOR_BGR2GRAY)
    ret1, threshold1 = cv2.threshold(gray1, 160, 255, cv2.THRESH_BINARY)
    nonzero1 = cv2.countNonZero(threshold1)
    total1 = gray1.shape[0] * gray1.shape[1]
    zero1 = total1 - nonzero1
    ratio1 = nonzero1 * 100 / float(total1)
    return ratio1
```

```python
source_dir = 'data/'
```

```python
file_names = os.listdir(source_dir)
```

```python
target_file = 'dataset'
os.mkdir(target_file)
os.mkdir(os.path.join(target_file, 'bad'))
os.mkdir(os.path.join(target_file, 'good'))
```

```python
for file in file_names:
    location = os.path.join(source_dir, file)
    target1 = data
    output = func1(location)
    print(file, output)
    if(output >= 92):
        shutil.copy(os.path.join(source_dir, file), os.path.join(target_file, 'good'))
    else:
        shutil.copy(os.path.join(source_dir, file), os.path.join(target_file, 'bad'))
```
